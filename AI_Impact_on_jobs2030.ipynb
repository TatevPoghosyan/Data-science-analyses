# üß© Feature Engineering

### **Skill‚Äì’•÷Ä’´ ’∞’°’¥’°’Ø÷Å’æ’°’Æ ’∞’°’ø’Ø’°’∂’´’∑**

```python
df_original = df.copy()
df_original.head()

skill_cols = [col for col in df.columns if "Skill_" in col]
df['Avg_Skill_Level'] = df[skill_cols].mean(axis=1)
```

’ç’ø’•’≤’Æ’æ’•÷Å ’∂’∏÷Ä ’∞’°’ø’Ø’°’∂’´’∑, ’∏÷Ä’® ’∂’•÷Ä’Ø’°’µ’°÷Å’∂’∏÷Ç’¥ ’ß ’°’∑’≠’°’ø’∏’≤’´ ’®’∂’§’∞’°’∂’∏÷Ç÷Ä ’∞’¥’ø’∏÷Ç’©’µ’∏÷Ç’∂’∂’•÷Ä’´ ’¥’°’Ø’°÷Ä’§’°’Ø’®÷â

### **‘±’∑’≠’°’ø’°’æ’°÷Ä’± / ’ì’∏÷Ä’±’°’º’∏÷Ç’©’µ’∏÷Ç’∂**





```python
df['Salary_per_Year_Exp'] = df['Average_Salary'] / (df['Years_Experience'] + 1)
```

## **Automation Risk Score**





```python
df['Automation_Risk_Score'] = (
    df['AI_Exposure_Index'] * df['Automation_Probability_2030']
)
```
